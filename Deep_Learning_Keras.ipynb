**Criação de um modelo de rede neural com Keras**

Criaremos um modelo simples de rede neural para classificação de imagens usando o conjunto de dados MNIST, que contém imagens de dígitos escritos à mão.

```python
import numpy as np
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
#Carregar o conjunto de dados MNIST (já divido em treinamento e teste)
(x_train, y_train), (x_test, y_test) = mnist.load_data()

#Pré_processamento dos dados
x_train = x_train.reshape((60000, 28 * 28)) #REdimensionar para um vetor unidimensional
x_train = x_train.astype('float32') / 255.0 #Normalizar os valores de pixel para o intervalo [0, 1]

x_test = x_test.reshape((10000, 28 * 28))
x_test = x_test.astype('float32') / 255.0

#Converter os rótulos em categorias one-hot
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

#One-hot encode das classes de saída
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
#Agora vamos criar o modelo de Rede Neural

#Inicializar o modelo sequencial
model = Sequential()

#Adicionar a camada de entrada e a primeira camada oculta
model.add(Dense(units=128, activation='relu', input_shape=(28 * 28,)))

#Adcionar a camada oculta
model.add(Dense(units=64, activation='relu'))

#Adicionar a camada de saída
model.add(Dense(units=10, activation='softmax'))
```
